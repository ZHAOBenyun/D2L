```python
import torch
from torch import nn
from torch.nn import functional as F
from d2l import torch as d2l
```

```python
# 'num_channels' is the output channels
class Residual(nn.Module):
    def __init__(self, input_channels, num_channels, use_1x1conv=False, strides=1):
        super().__init__()
        # conv1 can be given stride
        self.conv1 = nn.Conv2d(input_channels, num_channels, kernel_size=3, padding=1, stride=strides)
        self.conv2 = nn.Conv2d(num_channels, num_channels, kernel_size=3, padding=1)
        # 1x1conv is optional, stride can be given to match H&W
        if use_1x1conv:
            self.conv3 = nn.Conv2d(input_channels, num_channels, kernel_size=1, stride=strides)
        else:
            self.conv3 = None
        self.bn1 = nn.BatchNorm2d(num_channels)  # two BN layer have different parameters to learn
        self.bn2 = nn.BatchNorm2d(num_channels)
        self.relu = nn.ReLU(inplace=True)       # 'inplace' means operate on input address in memory
        
    def forward(self, X):
        Y = self.relu(self.bn1(self.conv1(X)))
        Y = self.bn2(self.conv2(Y))
        if self.conv3:
            X = self.conv3(X)  # we do not need ReLU, since this is applying on X not Y
        Y += X
        return self.relu(Y)
```

### Make the Shape of Input and Output Same

```python
blk = Residual(3, 3)
X = torch.randn(4, 3, 6, 6)
Y = blk(X)
Y.shape
```

### Build ResNet Model

```python
stage1 = nn.Sequential(
                nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3),
                nn.BatchNorm2d(64), nn.ReLU(),
                nn.MaxPool2d(kernel_size=3, stride=2, padding=1)
                    )
```

```python
# 'num_residuals' is the number of residual blocks
def resnet_block(input_channels, num_channels, num_residuals, first_block=False):
    blk = []
    for i in range(num_residuals):
        if i == 0 and first_block==False:  
            # in a sequence of residual, the first block will halve the H&W
            blk.append(Residual(input_channels, num_channels, use_1x1conv=True, strides=2))
        else:   
            # the rest blocks will not halve the H&W
            blk.append(Residual(num_channels, num_channels))
    return blk
```

```python
stage2 = nn.Sequential(*resnet_block(64, 64, 2, first_block=True))
        # first block do not need halve the H&W, since H&W have already been halved in stage 1
stage3 = nn.Sequential(*resnet_block(64, 128, 2))  # * means unzip the list; ** means unzip the dictionary
stage4 = nn.Sequential(*resnet_block(128, 256, 2))
stage5 = nn.Sequential(*resnet_block(256, 512, 2))
```

```python
net = nn.Sequential(
                stage1, stage2, stage3, stage4, stage5, 
                nn.AdaptiveAvgPool2d((1, 1)),
                nn.Flatten(), nn.Linear(512, 10)
                   ) 
```

### Show the ResNet

```python
# First stage is same
# Stage2 is same as stage1
# From stage3, double the channels and halve the H&W
X = torch.rand(size=(1, 1, 224, 224))
for layer in net:
    X = layer(X)
    print(layer.__class__.__name__, 'output shape:\t', X.shape)
```

```python
lr, num_epochs, batch_size = 0.05, 10, 32
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size, resize=224)
d2l.train_ch6(net, train_iter, test_iter, num_epochs, lr, d2l.try_gpu())
```
