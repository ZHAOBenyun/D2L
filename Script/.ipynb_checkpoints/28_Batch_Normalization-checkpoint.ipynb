```python
import torch
from torch import nn
from d2l import torch as d2l
```

## Batch Normalization is applied in All Samples on Feature Dimension, 
## which means the normalization is within the samples not across the samples.  
## When the batch size equals to 1, the BN cannot be used.

### Batch Normalization Step by Step

```python
# 'gamma & beta' is learnable
# 'moving_mean & moving_var' is global mean and var, not mini-batch mean and var
# 'eps' is a small num to ensure the denominator is not 0
# 'momentum' is to update 'moving_mean & moving_var'
def batch_norm(X, gamma, beta, moving_mean, moving_var, eps, momentum):
    if not torch.is_grad_enabled(): # 'is_grad_enabled' to determine whether the current mode 
                                    # is training or prediction mode
                                    # Here, this is the prediction mode
        X_hat = (X - moving_mean) / torch.sqrt(moving_var + eps) # torch.sqrt = calculate square root 
        # During the prediction, global mean and var are used
    else:
        assert len(X.shape) in (2,4)  # if X.shape is not 2 or 4, an error will come
        if len(X.shape) == 2:    # X.shape=2 means FC layer
            mean = X.mean(dim=0)               # get mean by row(dimension=0), result will be (n,1)
            var = ((X - mean)**2).mean(dim=0)  # get the var by row(dimension=0), result will be (n,1)
        else:                    # X.shape=4 means Conv2d layer
            mean = X.mean(dim=(0,2,3), keepdim=True)
                                               # get mean by channels(dimension=1), result will be (1,n,1,1)
            var = ((X - mean)**2).mean(dim=(0,2,3), keepdim=True)
                                               # get var by channels(dimension=1), result will be (1,n,1,1)
        X_hat = (X - mean) / torch.sqrt(var + eps)  
        # During the training, mean and var are used the result we calculated before
        moving_mean = momentum * moving_mean + (1 - momentum) * mean  # update the moving_mean
        moving_var = momentum * moving_var + (1 - momentum) * var     # update the moving_var
    Y = gamma * X_hat + beta  # Scale and shift
    return Y, moving_mean.data, moving_var.data
```

#### Define Batch Normalization layer

```python
class BatchNorm(nn.Module):
    def __init__(self, num_features, num_dims):
        super().__init__()
        if num_dims == 2:  # X.shape=2 means FC layer
            shape = (1, num_features)
        else:             # X.shape=4 means Conv2d layer
            shape = (1, num_features, 1, 1)
        self.gamma = nn.Parameter(torch.ones(shape))
        self.beta = nn.Parameter(torch.zeros(shape))
        self.moving_mean = torch.zeros(shape)
        self.moving_var = torch.ones(shape)
        
    def forward(self, X):
        if self.moving_mean.device != X.device:  # move 'moving_mean & moving_var' to GPU
            self.moving_mean = self.moving_mean.to(X.device)
            self.moving_var = self.moving_var.to(X.device)
        Y, self.moving_mean, self.moving_var = batch_norm(
                    X, self.gamma, self.beta, self.moving_mean, self.moving_var, eps=1e-5, momentum=0.9
                                                            )
        return Y
    
```

#### Apply BN layer on LeNet

net = nn.Sequential(
    nn.Conv2d(1, 6, kernel_size=5), BatchNorm(6, num_dims=4),
    nn.Sigmoid(), nn.MaxPool2d(kernel_size=2, stride=2),
    nn.Conv2d(6, 16, kernel_size=5), BatchNorm(16, num_dims=4),
    nn.Sigmoid(), nn.MaxPool2d(kernel_size=2, stride=2),
    nn.Flatten(), nn.Linear(16*4*4, 120), BatchNorm(120, num_dims=2),
    nn.Sigmoid(), nn.Linear(120, 84), BatchNorm(84, num_dims=2),
    nn.Sigmoid(), nn.Linear(84, 10)
    )

```python
# Adding the BN layer will make the map more flat, so network converges faster
lr, num_epochs, batch_size = 0.9, 10, 256
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
d2l.train_ch6(net, train_iter, test_iter, num_epochs, lr, d2l.try_gpu())
```

### Batch Normalization easily from PyTorch

```python
# PyTorch will adjust the num of dimension of BN layer automatically
net = nn.Sequential(
    nn.Conv2d(1, 6, kernel_size=5), 
    nn.BatchNorm2d(6),
    nn.Sigmoid(), nn.MaxPool2d(kernel_size=2, stride=2),
    nn.Conv2d(6, 16, kernel_size=5), 
    nn.BatchNorm2d(16),
    nn.Sigmoid(), nn.MaxPool2d(kernel_size=2, stride=2),
    nn.Flatten(), nn.Linear(16*4*4, 120), 
    nn.BatchNorm1d(120), 
    nn.Sigmoid(), nn.Linear(120, 84), 
    nn.BatchNorm1d(84),
    nn.Sigmoid(), nn.Linear(84, 10)
    )
# After Conv2d, we use BatchNorm'2d'; while after Linear, we use BatchNorm'1d'
```

```python
d2l.train_ch6(net, train_iter, test_iter, num_epochs, lr, d2l.try_gpu())
```
