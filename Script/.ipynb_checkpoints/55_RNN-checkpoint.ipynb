### Initialize the Model Params

```python
# just like one-hidden layer MLP
def get_params(vocab_size, num_hiddens, device):
    # through one_hot, we get the num_inputs as same as the length of vocab_size
    num_inputs = num_outputs = vocab_size
    
    def normal(shape):
        return torch.randn(size=shape, device=device) * 0.01
    
    # hidden layer params
    W_xh = normal((num_inputs, num_hiddens))
    W_hh = normal((num_hiddens, num_hiddens)) # the last time's hidden variables
    b_h = torch.zeros(num_hiddens, device=device)
    
    # output layer params
    W_hq = normal((num_hiddens, num_outputs))
    b_q = torch.zeros(num_outputs, device=device)
    
    # attach gradients
    params = [W_xh, W_hh, b_h, W_hq, b_q]
    for param in params:
        param.requires_grad_(True)
    return params
```

```python
def init_rnn_state(batch_size, num_hiddens, device):
    '''initialize the hidden state'''
    return (torch.zeros((batch_size, num_hiddens), device=device), )  # this return variable is a tuple (xxxx, )
```

```python
# 'inputs' is from X0 to Xt which is 3D tensor, we can see from BOX5
# 'state' is the initialized state
def rnn(inputs, state, params):
    '''just like a forward function in MLP'''
    W_xh, W_hh, b_h, W_hq, b_q = params
    H, = state
    outputs = []
    
    for X in inputs:  # iterate this 3D tensor from the first dimension which is from X0 to Xt
        H = torch.tanh(torch.mm(X, W_xh) 
                       + torch.mm(H, W_hh)   # this is the difference between MLP and RNN
                       + b_h)
        Y = torch.mm(H, W_hq) + b_q
        outputs.append(Y)
    return torch.cat(outputs, dim=0), (H, )   # show the output and the current hidden state
    # the output is a 2D tensor, shape is (batch_size * num of Xt, vocab_size)
```

### Define a Class to Combine all Functions together

```python
class RNNModelScratch:
    '''a RNN model from scratch'''
    def __init__(self, vocab_size, num_hiddens, device, get_params, init_state, forward_fn):
        self.vocab_size, self.num_hiddens = vocab_size, num_hiddens
        self.params = get_params(vocab_size, num_hiddens, device)
        self.init_state, self.forward_fn = init_state, forward_fn
        
    # this 'X' shape is (batch_size, num_steps)    
    def __call__(self, X, state):   # ***__call__ is to make the instance of this class used like a function
        # this 'X' shape is (num_steps, batch_size, vocab_size)
        X = F.one_hot(X.T, self.vocab_size).type(torch.float32)
        return self.forward_fn(X, state, self.params)
    
    def begin_state(self, batch_size, device):
        return self.init_state(batch_size, self.num_hiddens, device)
```

### Instantiate this RNN Class

```python
num_hiddens = 512
net = RNNModelScratch(len(vocab), num_hiddens, d2l.try_gpu(), get_params, init_rnn_state, rnn)
```

### Prediction

```python
def predict_ch8(prefix, num_preds, net, vocab, device):  #@save
    """Generate new characters following the `prefix`."""
    state = net.begin_state(batch_size=1, device=device)  # initialize the hidden state 
    outputs = [vocab[prefix[0]]]                  # save the subscript of words in vocab
    
    # outputs[-1] is the neartest word we predict, then we use it to predict the next word
    get_input = lambda: torch.tensor([outputs[-1]], device=device).reshape((1, 1))
    
    for y in prefix[1:]:  # use gt to warm up the period
        _, state = net(get_input(), state)
        # so at this time, we use gt prefix y
        outputs.append(vocab[y])   
    
    for _ in range(num_preds):  # we need 'num_preds' word, so we do 'num_preds' loop
        y, state = net(get_input(), state)
        # get the max y then turn it to int
        outputs.append(int(y.argmax(dim=1).reshape(1)))
    # turn the index into token,
    # use empty string to save all tokens, then make a full sentence
    return ''.join([vocab.idx_to_token[i] for i in outputs])
```

### Gradient Clip

(**$$\mathbf{g} \leftarrow \min\left(1, \frac{\theta}{\|\mathbf{g}\|}\right) \mathbf{g}.$$**)

```python
def grad_clipping(net, theta):
    if isinstance(net, nn.Module):
        params = [p for p in net.parameters() if p.requires_grad]
    else:
        params = net.params  # get all layers params together
        
    # square all params' grad, then get the sum of them
    # next, we sum all layers
    norm = torch.sqrt(sum(torch.sum((p.grad ** 2)) for p in params))
    if norm > theta:
        for param in params:
            param.grad[:] *= theta / norm
```

### Train one epoch

```python
# "use_random_iter" means the ith sample in last mini-batch has no relations with current ith sample
def train_epoch_ch8(net, train_iter, loss, updater, device, use_random_iter):
    state, timer = None, d2l.Timer()
    metric = d2l.Accumulator(2)
    
    for X, Y in train_iter:
        if state is None or use_random_iter:
            # initialize 'state' when either it is the first iteration which is none
            # or using random sampling
            # the reason is that the last sequence and current sequence are not continuous, 
            # so we cannot use the last state on current state
            state = net.begin_state(batch_size=X.shape[0], device=device)
        else:
            # if the last sequence and current sequence are continuous 
            # which means the ith sample in last mini-batch has relations with current ith sample
            # then we do not clear the last state, only detach them.
            if isinstance(net, nn.Module) and not isinstance(state, tuple):
                # when use .detach(), we drop all the computation graphs before this step
                # we only care about the gradient from here on out
                state.detach_()  
            else:
                for s in state:
                    s.detach_()
        
        y = Y.T.reshape(-1)
        X, y = X.to(device), y.to(device)
        y_hat, state = net(X, state)
        l = loss(y_hat, y.long()).mean()
        if isinstance(updater, torch.optim.Optimizer):
            updater.zero_grad()
            l.backward()
            grad_clipping(net, 1)
            updater.step()
        else:
            l.backward()
            grad_clipping(net, 1)  # use gradient clipping to prevent graident explosion
            updater(batch_size=1)
        metric.add(l * y.numel(), y.numel())
    # metric[0] / metric[1] is average cross-entropy loss
    # exp(metric[0] / metric[1]) is perplexity
    return math.exp(metric[0] / metric[1]), metric[1] / timer.stop()
```

### Train multiple Epochs

```python
def train_ch8(net, train_net, vocab, lr, num_epochs, device, use_random_iter=False):
    loss = nn.CrossEntropyLoss()
    animator = d2l.Animator(xlabel='epoch', ylabel='perplexity', legend=['train'], xlim=[10, num_epochs])
    
    if isinstance(net, nn.Module):
        updater = torch.optim.SGD(net.parameters(), lr)
    else:
        updater = lambda batch_size: d2l.sgd(net.params, lr, batch_size)
    
    predict = lambda prefix: predict_ch8(prefix, 50, net, vocab, device)
    
    for epoch in range(num_epochs):
        ppl, speed = train_epoch_ch8(net, train_iter, loss, updater, device, use_random_iter)
        if (epoch + 1) % 10 == 0:
            print(predict('time traveller'))
            animator.add(epoch + 1, [ppl])
    print(f'perplexity {ppl:.1f}, {speed:.1f} tokens/sec on {str(device)}')
    print(predict('time traveller'))
    print(predict('traveller'))
```

```python
num_epochs, lr = 500, 0.5
train_ch8(net, train_iter, vocab, lr, num_epochs, d2l.try_gpu())
```

```python
# when we use random sampling
train_ch8(net, train_iter, vocab, lr, num_epochs, d2l.try_gpu(), use_random_iter=True)
```

## RNN easily from PyTorch

```python
# num_steps means the length of sequence we read at one time
batch_size, num_steps = 32, 35
train_iter, vocab = d2l.load_data_time_machine(batch_size, num_steps)
```

### Define the Model

```python
num_hiddens = 256
rnn_layer = nn.RNN(len(vocab), num_hiddens)
```

```python
# initialize state
state = torch.zeros((1, batch_size, num_hiddens))
state.shape
```

```python
X = torch.rand(size=(num_steps, batch_size, len(vocab)))
Y, state_new = rnn_layer(X, state)

# Y is the hidden layer's shape
Y.shape, state_new.shape
```

```python
class RNNModel(nn.Module):
    def __init__(self, rnn_layer, vocab_size, **kwargs):
        super(RNNModel, self).__init__(**kwargs)
        self.rnn = rnn_layer
        self.vocab_size = vocab_size
        self.num_hiddens = self.rnn.hidden_size
        
        # If the RNN is bidirectional, 'num_directions' should be 2, else it should be 1.
        if not self.rnn.bidirectional:
            self.num_directions = 1
            # because PyTorch does not give us output layer of RNN
            self.linear = nn.Linear(self.num_hiddens, self.vocab_size)
        else:
            self.num_directions = 2
            self.linear = nn.Linear(self.num_hiddens * 2, self.vocab_size)
            
    def forward(self, inputs, state):
        X = F.one_hot(inputs.T.long(), self.vocab_size)
        X = X.to(torch.float32)
        Y, state = self.rnn(X, state)
        # use a fully connected layer to change the 3D tensor Y into 2D tensor 
        # new shape is (batch_size * num_steps, num_hiddens)
        output = self.linear(Y.reshape((-1, Y.shape[-1])))
        return output, state
    
    def begin_state(self, device, batch_size=1):
        if not isinstance(self.rnn, nn.LSTM):
            return torch.zeros((self.num_directions * self.rnn.num_layers, 
                                batch_size, self.num_hiddens), device=device)
        else:
            return (
                torch.zeros((self.num_directions * self.rnn.num_layers,
                             batch_size, self.num_hiddens), device=device),
                torch.zeros((self.num_directions * self.rnn.num_layers,
                             batch_size, self.num_hiddens), device=device)
                   )
```
